```python
# @title Imports
import gc
import os
from datetime import datetime
from hashlib import md5
from io import BytesIO
from warnings import filterwarnings

import torch
from PIL import Image
from diffusers import (
    FlowMatchEulerDiscreteScheduler,
    FlowMatchHeunDiscreteScheduler,
    StableDiffusion3Pipeline,
)
from transformers import logging
```

```python
# @title Pipeline
class Loader:
    _instance = None

    def __new__(cls):
        if cls._instance is None:
            cls._instance = super().__new__(cls)
            cls._instance.pipe = None
        return cls._instance

    def _flush(self):
        gc.collect()
        torch.cuda.empty_cache()
        torch.cuda.ipc_collect()
        torch.cuda.reset_peak_memory_stats()
        torch.cuda.synchronize()

    def _load_pipeline(self, cpu_offload, tqdm, **kwargs):
        if self.pipe is None:
            try:
                print("Loading stabilityai/stable-diffusion-3-medium...")
                file = "https://huggingface.co/stabilityai/stable-diffusion-3-medium/blob/main/sd3_medium_incl_clips_t5xxlfp8.safetensors"
                self.pipe = StableDiffusion3Pipeline.from_single_file(file, **kwargs)
                if cpu_offload:
                    self.pipe.enable_model_cpu_offload()
                else:
                    self.pipe.to("cuda")
            except Exception as e:
                print(e)
                self.pipe = None
        if self.pipe is not None:
            self.pipe.set_progress_bar_config(disable=not tqdm)

    def load(self, scheduler, cpu_offload, tqdm):
        schedulers = {
            "Euler": FlowMatchEulerDiscreteScheduler,
            "Heun": FlowMatchHeunDiscreteScheduler,
        }
        scheduler_kwargs = {"shift": 3.0}
        pipe_kwargs = {
            "torch_dtype": torch.float16,
            "scheduler": schedulers[scheduler](**scheduler_kwargs),
        }
        self._load_pipeline(cpu_offload, tqdm, **pipe_kwargs)

        # error
        if self.pipe is None:
            return

        same_scheduler = isinstance(self.pipe.scheduler, schedulers[scheduler])
        if not same_scheduler:
            print(f"Switching to {scheduler}...")
            self.pipe.scheduler = schedulers[scheduler](**scheduler_kwargs)
```

```python
# @title Inference
def main():
    global \
        SCHEDULER, \
        PROMPT, \
        NEGATIVE_PROMPT, \
        OUT_DIR, \
        SEED, \
        WIDTH, \
        HEIGHT, \
        GUIDANCE_SCALE, \
        INFERENCE_STEPS, \
        CPU_OFFLOAD, \
        PROGRESS

    if not torch.cuda.is_available():
        raise RuntimeError("CUDA not available")

    if SEED is None or SEED < 0:
        seed = int(datetime.now().timestamp() * 1e6) % 2**64
    else:
        seed = SEED

    loader = Loader()
    loader.load(SCHEDULER, CPU_OFFLOAD, PROGRESS)

    if loader.pipe is None:
        return

    pipe = loader.pipe

    try:
        image = pipe(
            width=WIDTH,
            height=HEIGHT,
            prompt=PROMPT,
            negative_prompt=NEGATIVE_PROMPT,
            guidance_scale=GUIDANCE_SCALE,
            num_inference_steps=INFERENCE_STEPS,
            generator=torch.Generator(device=pipe.device).manual_seed(seed),
        ).images[0]
    except Exception as e:
        print(e)
        return

    image_file = save_image(image, OUT_DIR)
    print(image_file)
    display(Image.open(image_file))


# save image using content-based hash
def save_image(image, out_dir="."):
    image_bytes = BytesIO()
    image.save(image_bytes, format="PNG")
    image_bytes.seek(0)
    image_hash = md5(image_bytes.read()).hexdigest()
    filename = f"{out_dir}/{image_hash}.png"
    image.save(filename)
    return filename
```

```python
# fmt: off
SCHEDULER = "Euler"  # @param {type: "string"} ["Euler", "Heun"]
PROMPT = "Majestic Machu Picchu, set against a backdrop of towering mountains, breathtaking, high detail, landscape"  # @param {type: "string"}
NEGATIVE_PROMPT = "ugly"  # @param {type: "string"}
OUT_DIR = "/content"  # @param {type: "string"}
SEED = 479081197198597  # @param {type: "integer", min: -1, max: 18446744073709551615}
WIDTH = 1344  # @param {type: "slider", min: 768, max: 1536, step: 64}
HEIGHT = 768  # @param {type: "slider", min: 768, max: 1536, step: 64}
GUIDANCE_SCALE = 5  # @param {type: "slider", min: 0, max: 15, step: 0.1}
INFERENCE_STEPS = 30  # @param {type: "slider", min: 1, max: 50}
CPU_OFFLOAD = False  # @param {type: "boolean"}
PROGRESS = False  # @param {type: "boolean"}
main()
```
